```python
# import pd and plt

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read csv file:heart-disease
df=pd.read_csv("/content/heart-disease.csv")
df
```

```python
#take the data that is over 50
over_50=df[df["age"]>50]
over_50
```

```python
#plot the data
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax.set(title="Heart disease and sholesterol",xlabel="Age",ylabel="Chol")
ax.legend(*scatter.legend_elements(),title="Target")
ax.axhline(over_50["chol"].mean(),linestyle="--")
```

### create figure (plot) with 2 axes

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,6))
scatter=ax0.scatter(x=over_50["age"],y=over_50["trestbps"],c=over_50["target"])
ax0.set(title="Heart disease and trestbps",xlabel="Age", ylabel="Trestbps")
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(over_50["trestbps"].mean(),linestyle="--")

scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.set(xlabel="Age",ylabel="Thalach", title="Heart disease and thalach")
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(over_50["thalach"].mean(), linestyle="--")

#title the figure
fig.suptitle("Heart Disease Analysis")
```
